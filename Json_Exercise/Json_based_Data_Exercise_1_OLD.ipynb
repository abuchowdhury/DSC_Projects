```python
import pandas as pd

import json
from pandas.io.json import json_normalize

# =============================================================================
# JSON exercise
# Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
# 1. Find the 10 countries with most projects
# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')
# 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
# 
# =============================================================================


# =============================================================================
# Q1. Find the 10 countries with most projects.
# =============================================================================

# We have to find unique project_name by country
# Display top 10 countries with most projects

# Print contry name  
c=sample_json_df['countryname']
print(c)

per_projects_unique = per_country_projects['project_name'].value_counts()
print('Display unique project')
per_projects_unique.head(10)

# Answers to Questions 1

print('Top 10 countries with number of unique projects:')
per_country_projects_unique = per_country_projects['countryname'].value_counts()
per_country_projects_unique.head(10)

```

```python
# =============================================================================
# Q2. Find the top 10 major project themes (using column 'mjtheme_namecode')
# 
# =============================================================================

mjtheme=sample_json_df['mjtheme_namecode']
print(mjtheme.head(11))
# First we need to create a themes dataframe using column 'mjtheme_namecode' and display it
# mjtheme_namecode consist of code and name {'code': '8', 'name': 'Human development'}

themes = pd.DataFrame(columns=['code', 'name'])
for theme in sample_json_df.mjtheme_namecode:
    themes = themes.append(json_normalize(theme))
themes.reset_index(drop=True, inplace=True)

themes.head(25)

# display name and code seperately 
themes['name']
themes['code']

# Display theme name

themes_name = themes.name
print(themes_name.head(30))

# Count project unique theme_names

themes_name_values = themes.name.values
print(themes_name_values)

# Answer to Question 2

# Count project unique theme_names

theme_name_value_counts = themes.name.value_counts()
print(theme_name_value_counts)

# display the top 10 number of project themes

print('Top 10 major project themes for World_bank:')
theme_name_value_counts.head(10)
```

```python
# =============================================================================
# Q3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
# =============================================================================

# Display theme name

themes_name = themes.name
print(themes_name.head(20))

# Display theme name with counts to see missing theme name value_counts

themes_name_value_counts = themes.name.value_counts()
print(themes_name_value_counts)

# Current count of missing entries
print('Count of missing name entries:', len(themes[themes['name'] == '']))


# Create a theme name dictionary to map theme codes to theme names
# Display dictionary
theme_name_dict = {}

for theme in themes.itertuples():
    if theme[2] != '':  # theme name value not null
        theme_name_dict[theme[1]] = theme[2]
        
theme_name_dict

# Test both of these below will have same value
theme_name_dict[theme[1]]
theme[2]




# We need to fill in missing theme names using the theme name dictionary (theme_name_dict)
for theme in themes.itertuples():
    if theme[2] == '': # theme name value is null
        # missing theme name value will be updated based on matching code (theme[0])
        themes.set_value(theme[0], 'name', theme_name_dict[theme[1]])
        
# After fill in missing theme names, make sure there are no more missing entries
print('Count of missing name entries:', len(themes[themes['name'] == '']))

# Display theme name

themes_name = themes.name
print(themes_name.head(10))
```
